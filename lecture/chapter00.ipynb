```python
from sklearn.datasets import load_iris, load_wine
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()  # 딕셔너리
iris.keys()
# iris.values()
# iris.items()
```

## iris 연습문제

iris 사이킷런 데이터는 넘파이 배열임

```python
# iris 데이터를 판다스의 테이터프레임으로 변경해보세요
# 가장 마지막 열은 target으로 설정하세요

import pandas as pd

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris['target']
df
```

왼쪽 인덱스는 판다스가 자동으로 붙여준거

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis=1), df['target'], random_state=42)

# df에서 target열을 제거한 입력변수(feature)만 추출하여 즉, X
# df['target']이 타깃변수(정답). 즉, y
```

```python
X_train, X_test, y_train, y_test
```

### knn 분류 모델 가져오기

fit()메서드

predict()메서드

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print(knn.score(X_test, y_test))
```

```python
from sklearn.datasets import load_iris, load_wine
from sklearn.model_selection import train_test_split
```

## iris Ïó∞ÏäµÎ¨∏Ï†ú

iris ÏÇ¨Ïù¥ÌÇ∑Îü∞ Îç∞Ïù¥ÌÑ∞Îäî ÎÑòÌååÏù¥ Î∞∞Ïó¥ÏûÑ

```python
# iris Îç∞Ïù¥ÌÑ∞Î•º ÌåêÎã§Ïä§Ïùò ÌÖåÏù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Î≥ÄÍ≤ΩÌï¥Î≥¥ÏÑ∏Ïöî
# Í∞ÄÏû• ÎßàÏßÄÎßâ Ïó¥ÏùÄ targetÏúºÎ°ú ÏÑ§Ï†ïÌïòÏÑ∏Ïöî

import pandas as pd

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris['target']
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis=1), df['target'], random_state=42)

# dfÏóêÏÑú targetÏó¥ÏùÑ Ï†úÍ±∞Ìïú ÏûÖÎ†•Î≥ÄÏàò(feature)Îßå Ï∂îÏ∂úÌïòÏó¨ Ï¶â, X
# df['target']Ïù¥ ÌÉÄÍπÉÎ≥ÄÏàò(Ï†ïÎãµ). Ï¶â, y
```

### knn Î∂ÑÎ•ò Î™®Îç∏ Í∞ÄÏ†∏Ïò§Í∏∞

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print(knn.score(X_test, y_test))
```
